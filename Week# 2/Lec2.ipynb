### Image Resizing

```python
img = cv2.imread('./Fig0219(rose1024).tif', 0) #reading image shape-->1024,1024

#result shape-->1024/2=512,1024/2=512
result512 = np.zeros((img.shape[0]//2, img.shape[1]//2), dtype=np.uint8) 
result256 = np.zeros((img.shape[0]//4, img.shape[1]//4), dtype= np.uint8)
result128 = np.zeros((img.shape[0]//8, img.shape[1]//8), dtype= np.uint8)
result64 = np.zeros((img.shape[0]//16, img.shape[1]//16), dtype= np.uint8)
result32 = np.zeros((img.shape[0]//32, img.shape[1]//32), dtype= np.uint8)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if i % 2 == 0 and j % 2==0:
            result512[i//2,j//2] = img[i,j]
            result256[i//4, j//4] = img[i,j]
            result128[i//8, j//8] = img[i,j]
            result64[i//16, j//16] = img[i,j]
            result32[i//32, j//32] = img[i,j]
            

shape = img.shape
cv2.imshow(str(shape), img)
shape = result512.shape
cv2.imshow(str(shape), result512)
shape = result256.shape
cv2.imshow(str(shape), result256)
shape = result128.shape
cv2.imshow(str(shape), result128)
shape = result64.shape
cv2.imshow(str(shape), result64)
shape = result32.shape
cv2.imshow(str(shape), result32)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Fig0219(rose1024).tif', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (512,512))
plane8 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane7 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane6 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane5 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane4 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane3 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane2 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
plane1 = np.zeros((img.shape[0], img.shape[1]), dtype = 'uint8')
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i,j]>127:
            plane8[i,j] = img[i,j]
        elif img[i,j]<=127 and img[i,j]>=64:
            plane7[i,j] = img[i,j]
        elif img[i,j] < 64 and img[i,j]>=32:
            plane6[i,j] = img[i,j]
        elif img[i,j] < 32 and img[i,j]>=16:
            plane5[i,j] = img[i,j]
        elif img[i,j] < 16 and img[i,j]>=8:
            plane4[i,j] = img[i,j]
        elif img[i,j] < 8 and img[i,j]>=4:
            plane3[i,j] = img[i,j]
        elif img[i,j] <4 and img[i,j]>=2:
            plane2[i,j] = img[i,j]
        else:
            plane1[i,j] = img[i,j]
cv2.imshow('Original', img)
cv2.imshow('Plane8', plane8)
cv2.imshow('Plane7', plane7)
cv2.imshow('Plane6', plane6)
cv2.imshow('Plane5', plane5)
cv2.imshow('Plane4', plane4)
cv2.imshow('Plane3', plane3)
cv2.imshow('Plane2', plane2)
cv2.imshow('Plane1', plane1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Up sampling /Interpolation

```python
img = cv2.imread('./Fig0241(c)(einstein high contrast).tif')
frame = np.zeros((1024,1024,3), dtype = 'uint8')

a = 0; b = 0;
for i in range(frame.shape[0]//2):
    a = i*2
    for j in range(frame.shape[1]//2):
        b =j*2
        frame[a,b] = img[i,j]
cv2.imshow('Zoomed', frame)
# row interpolation
print(frame[0,1])
for i in range(0,1024,2):
    for j in range(1,1024, 2):
        p = j-1
        f = j+1
        if f == 1024:
            frame[i,j] = frame[i,p]/2
        else:
            frame[i,j] = frame[i,p]/2+frame[i, f]/2
print(frame[0,1])
cv2.imshow('Rowwise', frame)
#columns interpolation
print(frame[1,0])
for i in range (1,1024,2):
    p = i-1
    f = i+1
    for j in range(0,1024,2):
        if f == 1024:
            frame[i,j] = frame[p,j]/2
        else:
            frame[i, j] = frame[p,j]/2 + frame[f,j]/2
print(frame[1,0])
cv2.imshow('orginal', img)
cv2.imshow('Col', frame)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
